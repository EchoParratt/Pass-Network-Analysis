```python
events = sb.events(match_id = 3869685)
```

```python
print(events.columns)
```

As you can see we have now accessed all the event data we need from Argentina vs France

Next let's create a new dataset that focuses on the tactics and type columns in events

```python
tactics = events[events['tactics'].isnull() == False]
tactics = tactics[['tactics', 'team', 'type']]
```

```python
tactics
```

```python
tactics = tactics[tactics['type'] == 'Starting XI']
tactics_Fra = tactics[tactics['team'] == 'France']
tactics_Arg = tactics[tactics['team'] == 'Argentina']
tactics_Fra = tactics_Fra['tactics']
tactics_Arg = tactics_Arg['tactics']
```

```python
dict_Fra = tactics_Fra[1]['lineup']
dict_Arg = tactics_Arg[0]['lineup']
```

```python
lineup_Fra = pd.DataFrame.from_dict(dict_Fra)
lineup_Fra
```

```python
lineup_Arg = pd.DataFrame.from_dict(dict_Arg)
lineup_Arg
```

Now we have made it so both Argentina and France both have their own dataframe and it only includes the starting 11

```python
Fra_players = {}
for i in range(len(lineup_Fra)):
    key = lineup_Fra.player[i]['name']
    val = lineup_Fra.jersey_number[i]
    Fra_players[key] = str(val)
print(Fra_players)
```

```python
Arg_players = {}
for i in range(len(lineup_Arg)):
    key = lineup_Arg.player[i]['name']
    val = lineup_Arg.jersey_number[i]
    Arg_players[key] = str(val)
print(Arg_players)
```

Created two dictionaries each having the player name as the key and their jersey number as the value

Next let's extract the relevant data we need from events

```python
events_pass = events[['minute', 'second', 'team', 'type', 'location', 'pass_end_location', 'pass_outcome', 'player']]
```

```python
events_pass.head(10)
```

```python
pass_events_Fra = events_pass[events_pass['team'] == 'France']
pass_events_Arg = events_pass[events_pass['team'] == 'Argentina']
```

```python
pass_events_Fra.head(10)
```

```python
pass_events_Arg.head(10)
```

```python
pn_events_Fra = pass_events_Fra[pass_events_Fra['type'] == 'Pass']
pn_events_Arg = pass_events_Arg[pass_events_Arg['type'] == 'Pass']
```

```python
pn_events_Fra.head(10)
```

```python
pn_events_Arg.head(10)
```

Here we are adding two extra columns pass_maker and pass_receiver

```python
pn_events_Fra['pass_maker'] = pn_events_Fra['player']
pn_events_Fra['pass_receiver'] = pn_events_Fra['player'].shift(-1)
```

```python
pn_events_Fra.head(10)
```

```python
pn_events_Arg['pass_maker'] = pn_events_Arg['player']
pn_events_Arg['pass_receiver'] = pn_events_Arg['player'].shift(-1)
```

```python
pn_events_Arg.head(10)
```

```python
pn_events_Fra = pn_events_Fra[pn_events_Fra['pass_outcome'].isnull() == True].reset_index()
pn_events_Arg = pn_events_Arg[pn_events_Arg['pass_outcome'].isnull() == True].reset_index()
```

Now that our data has been filtered and modified a little, we are going to focus on the fact that we are only using players
in the starting 11 in our pass network. To do this we must remove the pass events that take place after the first subsitution of both teams

```python
sub_Fra = pass_events_Fra[pass_events_Fra['type'] == 'Substitution']
sub_Arg = pass_events_Arg[pass_events_Arg['type'] == 'Substitution']
```

```python
sub_Fra
```

```python
sub_Arg
```

```python
sub_Fra_minute = np.min(sub_Fra['minute'])
sub_Fra_minute_data = sub_Fra[sub_Fra['minute'] == sub_Fra_minute]
sub_Fra_second = np.min(sub_Fra_minute_data['second'])
print("minute = ", sub_Fra_minute, "second = ", sub_Fra_second)
```

```python
sub_Arg_minute = np.min(sub_Arg['minute'])
sub_Arg_minute_data = sub_Arg[sub_Arg['minute'] == sub_Arg_minute]
sub_Arg_second = np.min(sub_Arg_minute_data['second'])
print("minute = ", sub_Arg_minute, "second = ", sub_Arg_second)
```

```python
pn_events_Fra = pn_events_Fra[(pn_events_Fra['minute'] <= sub_Fra_minute)]
pn_events_Arg = pn_events_Arg[(pn_events_Arg['minute'] <= sub_Arg_minute)]
```

Next we will split the location column and the pass_end_location column into two columns respectively to represent the xy coordinates

```python
Loc = pn_events_Fra['location']
Loc = pd.DataFrame(Loc.to_list(), columns=['pass_maker_x', 'pass_maker_y'])

Loc_end = pn_events_Fra['pass_end_location']
Loc_end = pd.DataFrame(Loc_end.to_list(), columns=['pass_receiver_x', 'pass_receiver_y'])

pn_events_Fra['pass_maker_x'] = Loc['pass_maker_x']
pn_events_Fra['pass_maker_y'] = Loc['pass_maker_y']
pn_events_Fra['pass_receiver_x'] = Loc_end['pass_receiver_x']
pn_events_Fra['pass_receiver_y'] = Loc_end['pass_receiver_y']

pn_events_Fra = pn_events_Fra[['minute', 'second', 'team', 'type', 'pass_outcome',
                               'player', 'pass_maker', 'pass_receiver', 'pass_maker_x',
                               'pass_maker_y', 'pass_receiver_y', 'pass_receiver_x']]
```

```python
pn_events_Fra.head(10)
```

```python
Loc = pn_events_Arg['location']
Loc = pd.DataFrame(Loc.to_list(), columns=['pass_maker_x', 'pass_maker_y'])

Loc_end = pn_events_Arg['pass_end_location']
Loc_end = pd.DataFrame(Loc_end.to_list(), columns=['pass_receiver_x', 'pass_receiver_y'])

pn_events_Arg['pass_maker_x'] = Loc['pass_maker_x']
pn_events_Arg['pass_maker_y'] = Loc['pass_maker_y']
pn_events_Arg['pass_receiver_x'] = Loc_end['pass_receiver_x']
pn_events_Arg['pass_receiver_y'] = Loc_end['pass_receiver_y']

pn_events_Arg = pn_events_Arg[['minute', 'second', 'team', 'type', 'pass_outcome',
                               'player', 'pass_maker', 'pass_receiver', 'pass_maker_x',
                               'pass_maker_y', 'pass_receiver_y', 'pass_receiver_x']]
```

```python
pn_events_Arg.head(10)
```

```python
pn_events_Fra.reset_index(inplace=True)
```

```python
pn_events_Arg.reset_index(inplace=True)
```

Next we will take the average location of each player and the number of passes they completed

```python
average_loc_Fra = pn_events_Fra.groupby('pass_maker').agg({'pass_maker_x':['mean'],
                                                           'pass_maker_y':['mean','count']})
```

```python
average_loc_Arg = pn_events_Arg.groupby('pass_maker').agg({'pass_maker_x':['mean'],
                                                           'pass_maker_y':['mean','count']})
```

```python
average_loc_Fra
```

```python
average_loc_Arg
```

```python
average_loc_Fra.columns = ['pass_maker_x', 'pass_maker_y', 'count']
average_loc_Fra
```

```python
average_loc_Arg.columns = ['pass_maker_x', 'pass_maker_y', 'count']
average_loc_Arg
```

Now let's try to figure out how many times a unique player A passed the ball to another unique player

```python
pass_Fra = pn_events_Fra.groupby(['pass_maker', 'pass_receiver']).index.count().reset_index()
pass_Fra
```

```python
pass_Arg = pn_events_Arg.groupby(['pass_maker', 'pass_receiver']).index.count().reset_index()
pass_Arg
```

```python
pass_Arg.rename(columns = {'index': 'number_of_passes'}, inplace = True)
```

```python
pass_Fra.rename(columns = {'index': 'number_of_passes'}, inplace = True)
```

Now we will merge the datasets we have created pass_Fra and average_loc_Fra and pass_Arg and average_loc_Arg

```python
pass_Fra = pass_Fra.merge(average_loc_Fra, left_on = 'pass_maker', right_index = True)
```

```python
pass_Arg = pass_Arg.merge(average_loc_Arg, left_on = 'pass_maker', right_index = True)
```

```python
pass_Fra.head(10)
```

```python
pass_Arg.head(10)
```

```python
pass_Fra = pass_Fra.merge(average_loc_Fra, left_on = 'pass_receiver', right_index = True, suffixes = ['', '_receipt'])
pass_Fra.rename(columns = {'pass_maker_x_receipt': 'pass_receiver_x',
                           'pass_maker_y_receipt': 'pass_receiver_y',
                           'count_receipt': 'number_of_passes_received'}, inplace = True)
pass_Fra = pass_Fra[pass_Fra['pass_maker'] != pass_Fra['pass_receiver']].reset_index()
```

```python
pass_Arg = pass_Arg.merge(average_loc_Arg, left_on = 'pass_receiver', right_index = True, suffixes = ['', '_receipt'])
pass_Arg.rename(columns = {'pass_maker_x_receipt': 'pass_receiver_x',
                           'pass_maker_y_receipt': 'pass_receiver_y',
                           'count_receipt': 'number_of_passes_received'}, inplace = True)
pass_Arg = pass_Arg[pass_Arg['pass_maker'] != pass_Arg['pass_receiver']].reset_index()
```

```python
pass_Fra.head(10)
```

```python
pass_Arg.head(10)
```

Now that we have all the data we need we can now create our pass network

```python
pitch = pitch = Pitch(pitch_color = 'black', line_color = 'white', stripe = False,
                        goal_type = 'box', label = False,
                      axis = True, tick = False)
fig, ax = pitch.draw()
arrows = pitch.arrows(pass_Fra.pass_maker_x, pass_Fra.pass_maker_y,
                      pass_Fra.pass_receiver_x, pass_Fra.pass_receiver_y, lw = 5,
                      color = '#002654', zorder = 1, ax = ax)
nodes = pitch.scatter(average_loc_Fra.pass_maker_x, average_loc_Fra.pass_maker_y,
                      s = 350, color = '#ED2939', edgecolor = 'white', linewidth = 1, alpha = 1, ax = ax)

for index, row in average_loc_Fra.iterrows():
    jersey_number = Fra_players.get(row.name, 'Not found')
    pitch.annotate(jersey_number, xy=(row.pass_maker_x, row.pass_maker_y),
                   c = 'white', va = 'center', ha = 'center', size = 10, ax = ax)
plt.title("Pass network for France against Argentina", size = 20)
plt.show()
```

```python
pitch = pitch = Pitch(pitch_color = 'black', line_color = 'white', stripe = False,
                        goal_type = 'box', label = False,
                      axis = True, tick = False)
fig, ax = pitch.draw()
arrows = pitch.arrows(120 - pass_Arg.pass_maker_x, pass_Arg.pass_maker_y,
                      120 - pass_Arg.pass_receiver_x, pass_Arg.pass_receiver_y, lw = 5,
                      color = 'white', zorder = 1, ax = ax)
nodes = pitch.scatter(120 - average_loc_Arg.pass_maker_x, average_loc_Arg.pass_maker_y,
                      s = 350, color = '#D5F3FE', edgecolor = 'white', linewidth = 1, alpha = 1, ax = ax)

for index, row in average_loc_Arg.iterrows():
    jersey_number = Arg_players.get(row.name, 'Not found')
    pitch.annotate(jersey_number, xy=(120 - row.pass_maker_x, row.pass_maker_y),
                   c = 'black', va = 'center', ha = 'center', size = 10, ax = ax)
plt.title("Pass network for Argentina against France", size = 20)
plt.show()
```
